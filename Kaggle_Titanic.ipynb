```python
#changing "Sex" into numerical data
dummy_sex = pd.get_dummies(df_train['Sex'])
dummy_sex.head()
```

```python
#changing "Embarked" into numerical data
dummy_embarked = pd.get_dummies(df_train['Embarked'])
dummy_embarked.head()
```

```python
#train_data_Merge 
df_merge = pd.concat([df_train, dummy_sex, dummy_embarked], axis = 1)
df_merge.head()
```

```python
#delete some columns from train data
df_train_new = df_merge.drop('Name', axis=1)
df_train_new = df_train_new.drop('Sex', axis=1)
df_train_new = df_train_new.drop('Cabin', axis=1)
df_train_new = df_train_new.drop('Ticket', axis=1)
df_train_new = df_train_new.drop('Embarked', axis=1)
df_train_new = df_train_new.drop('Age', axis=1)
df_train_new = df_train_new.drop('Fare', axis=1)
df_train_new.head()
```

```python
#changing "test_Sex" into numerical data
dummy_test_sex = pd.get_dummies(df_test['Sex'])
dummy_test_sex.head()
```

```python
#changing "test_Embarked" into numerical data
dummy_test_embarked = pd.get_dummies(df_test['Embarked'])
dummy_test_embarked.head()
```

```python
#test_data_Merge 
df_merge_test = pd.concat([df_test, dummy_test_sex, dummy_test_embarked], axis = 1)
df_merge_test.head()
```

```python
#delete some columns from test data
df_test_new = df_merge_test.drop(['Name','Sex', 'Cabin', 'Embarked'], axis=1)
df_test_new.head()
```

```python
print(df_test_new.shape)
```

```python
print(df_train_new.shape)
```

```python
# Labels are the values we want to predict
labels = np.array(df_train_new['Survived'])

# Remove the labels from the features
# axis 1 refers to the columns
features = df_train_new.drop('Survived', axis = 1)

features.head()
```

```python
print(features.shape)
```

```python
print(labels.shape)
```

```python
# Using Skicit-learn to split data into training and testing sets
from sklearn.model_selection import train_test_split
# Split the data into training and testing sets
train_features, test_features, train_labels, test_labels = train_test_split(features, labels, test_size = 0.3, random_state = 42)
```

```python
print('Training Features Shape:', train_features.shape)
print('Training Labels Shape:', train_labels.shape)
print('Testing Features Shape:', test_features.shape)
print('Testing Labels Shape:', test_labels.shape)
```

```python
train_features.info()
```

```python
# Construction of the model
random_forest = RandomForestClassifier(max_depth=30, n_estimators=30, random_state=42)
random_forest.fit(train_features, train_labels)

# Prediction
y_pred = random_forest.predict(test_features)

# accuracy
trainaccuracy_random_forest = random_forest.score(train_features, train_labels)
print('TrainAccuracy: {}'.format(trainaccuracy_random_forest))


accuracy_random_forest = accuracy_score(test_labels, y_pred)
print('Accuracy: {}'.format(accuracy_random_forest))
```
